```python
import pandas as pd
import numpy as np
```

# Problem 1

```python
df = pd.read_csv('Data/restaurant.csv')
df
```

```python
#'Humans will not wait in line for food if they are hungry.'

# -> Testing for whether humans are hungry. That is the positive for will not wait.

#True positive :     hungry -> Won't wait
#True negative : not hungry -> Will wait
#False negative: not hungry -> Won't wait
#False positive:     hungry -> Will wait

N= len(df)
TP = 2
TN = 1
FP = 5
FN = 4
ActualNo = TN + FP
ActualYes = FN + TP
PredictedNo = TN + FN
PredictedYes = FP + TP
Accuracy  = float( TP + TN) / N
Misclassification = float(FP + FN)/N
Recall = float(TP)/ActualYes
FalsePositive = float(FP)/ActualNo
Specificity = float(TN)/ActualNo
Precision = float(TP)/PredictedYes
Prevalence = float(ActualYes)/N
F1Score = 2*float((1/((1/Recall) + (1/Precision))))


print "Confusion Matrix"
print "N = %d     |   Predicted No |   Predicted Yes  |   Counts" % (N)
print "Actual: No |       %d        |       %d          |      %d " % (TN,FP,ActualNo)
print "Actual: Yes|       %d        |       %d          |      %d " % (FN,TP,ActualYes)
print "Counts     |       %d        |       %d          |     --- " % (PredictedNo, PredictedYes)

print "\nAccuracy Rate: \t\t%f " % (Accuracy)
print "Misclassification Rate: %f " % (Misclassification) 
print "Recall: \t\t%f " % (Recall) 
print "False Positive Rate: \t%f " % (FalsePositive) 
print "Specificity: \t\t%f " % (Specificity) 
print "Precision: \t\t%f " % (Precision) 
print "Prevalence: \t\t%f " % (Prevalence) 
print "F1 Score: \t\t%f " % (F1Score) 
```

# Problem 2

```python
animals = pd.read_csv('Data/zoo.csv')
animals
```

```python
#predators are venomous. 
#Predict predator if animal has venom, predict non-predator otherwise. 

# -> Testing for whether animal has venom. That is the positive for a predator.

#True positive :    venom  -> predator
#True negative : no venom  -> not predator
#False negative: no venom  -> predator
#False positive:    venom  -> not predator

venomous = animals.groupby(["venomous"]).get_group(1)
venomousPred = venomous.groupby(["predator"]).get_group(1)
venomousNotPred = venomous.groupby(["predator"]).get_group(0)
nonVenomous = animals.groupby(["venomous"]).get_group(0)
NonVenomousPred = nonVenomous.groupby(["predator"]).get_group(1)
NonVenomousNotPred = nonVenomous.groupby(["predator"]).get_group(0)

N= len(animals)
TP = len(venomousPred)
TN = len(NonVenomousNotPred)
FP = len(venomousNotPred)
FN = len(NonVenomousPred)
ActualNo = TN + FP
ActualYes = FN + TP
PredictedNo = TN + FN
PredictedYes = FP + TP
Accuracy  = float( TP + TN) / N
Misclassification = float(FP + FN)/N
Recall = float(TP)/ActualYes
FalsePositive = float(FP)/ActualNo
Specificity = float(TN)/ActualNo
Precision = float(TP)/PredictedYes
Prevalence = float(ActualYes)/N
F1Score = 2*float((1/((1/Recall) + (1/Precision))))


print "Confusion Matrix"
print "N = %d     |   Predicted No |   Predicted Yes  |   Counts" % (N)
print "Actual: No |       %d        |       %d          |      %d " % (TN,FP,ActualNo)
print "Actual: Yes|       %d        |       %d          |      %d " % (FN,TP,ActualYes)
print "Counts     |       %d        |       %d          |     --- " % (PredictedNo, PredictedYes)

print "\nAccuracy Rate: \t\t%f " % (Accuracy)
print "Misclassification Rate: %f " % (Misclassification) 
print "Recall: \t\t%f " % (Recall) 
print "False Positive Rate: \t%f " % (FalsePositive) 
print "Specificity: \t\t%f " % (Specificity) 
print "Precision: \t\t%f " % (Precision) 
print "Prevalence: \t\t%f " % (Prevalence) 
print "F1 Score: \t\t%f " % (F1Score) 

```
